### Penn Fudan

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import dl_toolbox.datamodules as datamodules
from pathlib import Path
import torchvision.transforms.v2 as v2
import torch
from torchvision import tv_tensors
import matplotlib.pyplot as plt
from torchvision.transforms.v2 import functional as F
import torchvision.utils as tv_utils
import pytorch_lightning as pl
from dl_toolbox.utils import show_detections

pl.seed_everything(2)

tf = v2.Compose(
    [
        v2.Resize(size=480,max_size=672),
        v2.RandomCrop(size=(672,672), pad_if_needed=True, fill=0),
        v2.ConvertBoundingBoxFormat('XYWH'),
        v2.SanitizeBoundingBoxes(),
    ]
)
 
dm = datamodules.PennFudan(
    data_path='/data',
    merge='all',
    train_tf=tf,
    val_tf=tf,
    test_tf=tf,
    batch_size=4,
    num_workers=4,
    pin_memory=False,
    epoch_steps=250
)

dm.prepare_data()
dm.setup(stage='fit')
```

```python
num_boxes = []
box_areas = []
for batch, batch_idx, dataloader_idx in dm.train_dataloader():
    batch = batch['sup']
    show_detections(batch['image'], batch['target'], class_list=dm.class_list)
    num = [int(b['labels'].shape[0]) for b in batch['target']]
    areas = [int(b[2]*b[3]) for t in batch['target'] for b in t['boxes']]
    num_boxes += num
    box_areas += areas
    break
    
print(num_boxes)
print(box_areas)
```
